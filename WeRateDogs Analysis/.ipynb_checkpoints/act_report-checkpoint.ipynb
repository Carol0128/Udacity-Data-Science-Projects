## WeRateDogs Analysis

[WeRateDogs](https://twitter.com/dog_rates) is a Twitter account that rates people's dogs with humorous comment about the dog. 

The objective of this project is to  wrangle the tweet archive of WeRateDogs, provided by Udacity, then eventually creating interesting and trustworthy analyses and visualizations.

The archive data has limited information about a tweet. To give more information about a particular tweet, we'll scrape more tweet info like the number of likes and retweets from the twitter account. 

Additionally, image predictions for the breed of a dog are provided.

Like most real world data, these datasets are messy and require to be properly wrangled before deriving any insights from them.

#### Data Wrangling

In this step, all the three datasets were gathered and loaded into the notebook using python functionalities. 

The python library, [tweepy](https://www.tweepy.org/), was very fundamental in scraping the additional information from Twitter API.

After gathering, quality and tidiness issues in the datasets were assessed using both programmatic and visual assessment, and cleaned accordingly.

Finally, the cleaned datasets were merged into one master dataframe and explored for insights.

#### Analysis and Visualization

The analysis aimed at answering the following questions:

1. What's the mean rating for the dogs?

2. What dog stage has the highest number of likes?

3. What's the most common dog name?

**Mean Rating**

WeRateDogs has a unique rating system where the rating numerator can be greater than the denominator. 

With the majority ratings between 11.3 and 13, the mean rating for the dogs is 11.63.
