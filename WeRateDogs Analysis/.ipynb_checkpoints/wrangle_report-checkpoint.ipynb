## WeRateDogs Data Wrangling Project

WeRateDogs is a Twitter account that rates people's dogs with humorous comment about the dog. The account reminds its followers that all dogs are good dogs.


<img src="/home/carol/dev/ALX-T/Udacity-Exercises/project2_extra_files/FV0AZc9WAAAqBVd.jpeg">


The goal of this project was wrangling the tweet archive of the twitter user [@dog_rates](https://twitter.com/dog_rates), then eventually creating interesting and trustworthy analyses and visualizations.

To give more information about a particular tweet, we scraped more information like the number of likes and retweets from the twitter account. Additionally, image predictions for the breed of dogs were provided.

The datasets used in this project are:

1. Twitter Archive Data :- Contains basic tweet data for over 2000 tweetw from WeRateDogs account

2. Tweet Json :- Additional information about the tweets gathered from Twitter's API

2. Image Predictions :- Image predictions for dog breeds


#### Data Gathering

The three datasets were gathered as elaborated below:

**Archive data**

This data was provided by udacity, as a csv file. Using pandas, it was read and loaded into the notebook for assessment and analysis

**Additional Information for tweets**

The archive data contains basic information about the tweets. To get additional information like number of likes and retweets, we used [tweepy](https://www.tweepy.org/) library to scrape the data from Twitter API.

**Image Predictions**

From the url provided by udacity, the image predictions was downloaded as a tsv file using [requests](https://requests.readthedocs.io/en/latest/) library.


#### Assessing Data

Each of the datasets were assessed visually and programmatically for any quality or tidiness issues.

The following issues were found in the data:-

***Quality Issues***

1. Incorrect datatype for timestamp

2. Retweets in the archive data

3. outliers in the rating numerators and denominators

4. Incorrect dog names like such, a, an, the, quite

5. Html tags in the source column

6. Duplicate urls in expanded_url column

7. Some values in the image predictions dataset are not images of dogs

8. Retweets in the scraped information

***Tidiness Issues***

1. Hyperlinks in the text column of archive data

2. Dog stages in separate columns

3. Have one table with all the information


#### Data Cleaning

Each of the issues listed above were then cleaned using various techniques.

The final clean datasets were merged into one master csv file, stored and analyzed for insights.

Some of the insights are:-

- The mean rating for the dogs is 12

- Puppers, which is an affectionate term for puppies, had the highest number of likes

- The most common names in this data were Lucy and Charlie.
